Load the dataset and display the first 15 rows

```python
import pandas as pd

df = pd.read_csv("Doctor_visit.csv")
print(df.head(15))
```

Display complete information about the columns of the dataset

```python
print(df.info())
```

Find out the total number of people based on their count of illness

```python
illness_count = df['No_of_illness'].value_counts().sort_index()
print(illness_count)
```

Visualize and analyze the maximum, minimum, and median income:

```python
import matplotlib.pyplot as plt

plt.hist(df['Income'], bins=20)
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Income Distribution')
plt.show()

print("Maximum Income:", df['Income'].max())
print("Minimum Income:", df['Income'].min())
print("Median Income:", df['Income'].median())
```

Find out the number of days of reduced activity of male and female separately due to illness:

```python
male_reduced_activity = df.loc[df['Gender'] == 'Male', 'Reduced_activity_days'].sum()
female_reduced_activity = df.loc[df['Gender'] == 'Female', 'Reduced_activity_days'].sum()
print("Male Reduced Activity Days:", male_reduced_activity)
print("Female Reduced Activity Days:", female_reduced_activity)
```

Visualize if there are any missing values in the dataset based on a heat map:

```python
import seaborn as sns

sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

Find out the correlation between variables in the given dataset:

```python
print(df.corr())
```

Analyze how the income of a patient affects the number of visits to the hospital:

```python
plt.scatter(df['Income'], df['No_of_visits'])
plt.xlabel('Income')
plt.ylabel('Number of Visits')
plt.title('Income vs Number of Visits')
plt.show()
```

Count and visualize the number of males and females affected by illness

```python
male_count = len(df[df['Gender'] == 'Male'])
female_count = len(df[df['Gender'] == 'Female'])
print("Male Count:", male_count)
print("Female Count:", female_count)

labels = ['Male', 'Female']
values = [male_count, female_count]
plt.bar(labels, values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

Visualize the percentage of people getting govt health Insurance due to low income, due to old age and also the percentage of people having private health insurance

```python
health_insurance = df['health_insurance'].value_counts()
print(health_insurance)
health_insurance.plot(kind='pie', autopct='%1.1f%%')
plt.title("Health Insurance Percentage")
plt.show()
```

Plot a horizontal bar chart to analyze the reduced days of activity due to illness based on Gender

```python
sns.catplot(x='Reduced_activity_days', y='Gender', kind='bar', data=df)

# Show the plot
plt.show()
```
