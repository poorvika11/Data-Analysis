#Toss a biased coin
For  pâ‰ q , we require a method to toss a biased coin. This is accomplished by the following function that generates  m  coin tosses with probability of heads equal to  p . Note that a value of 1 represents heads and 2 represents tails as before

After calculated manually probabilty coming as 0.251 and 0.252.

```python
import numpy as np
def biased(p, m):
  return 2-(np.random.rand(m) < p)
no_heads = 0   #variable for storing number of heads
p = 0.25
print(p)
for i in range(1000):
  if biased(p, 1) == 1:
    no_heads = no_heads + 1
print(no_heads/1000) #probability estimate by Monte Carlo
```

As you can see Monte carlo experiment verified as answers are coming same.
